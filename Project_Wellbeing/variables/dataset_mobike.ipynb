```python
import pandas as pd
import datetime
import time
```

```python
df_mobike = pd.read_pickle('./dataset/utseus-mobike-2016.pk')
```

```python
df_mobike.head()
```

```python
temps1 = datetime.time(7,0,0)
temps2 = datetime.time(8,0,0)
```

```python
values_start = {}
values_end = {}
```

```python
for orderid, start_time, end_time in zip(df_mobike['orderid'], df_mobike['start_time'], df_mobike['end_time']):
    start_date = datetime.datetime.strptime(start_time, '%Y-%m-%d %H:%M').time()
    end_date = datetime.datetime.strptime(end_time, '%Y-%m-%d %H:%M').time()
    
    if (start_date >= temps1) & (start_date <= temps2):
        values_start[orderid]=True
    else: values_start[orderid]=False
    
    if (end_date >= temps1) & (end_date >= temps2):
        values_end[orderid]=True
    else: values_end[orderid]=False
```

```python
df_mobike['start_time_in_range'] = df_mobike['orderid'].apply(lambda x: values_start[x])
df_mobike['end_time_in_range'] = df_mobike['orderid'].apply(lambda x: values_end[x])
```

```python
from shapely.geometry import Point
```

```python
geometry_start = [Point(xy) for xy in zip(df_mobike['start_location_x'].astype(float), df_mobike['start_location_y'].astype(float))]
geometry_end = [Point(xy) for xy in zip(df_mobike['end_location_x'].astype(float), df_mobike['end_location_y'].astype(float))]
```

```python
df_mobike['geometry_start'] = geometry_start
df_mobike['geometry_end'] = geometry_end
```

# Compute the number of mobike in each district

```python
df_district = pd.read_pickle('./dataset/utseus-cendus-shanghai-neighborhoods.pk')
df_district.head()
```

```python
df_mobike_per_discrit = df_district[['geometry']].copy()
```

```python
find_start = False
find_end = False
for index, row in df_mobike.iterrows():
    geom_start = row.geometry_start
    geom_end = row.geometry_end
    for index2, polygon in df_district['geometry'].items(): 
        if (not find_start & geom_start.within(polygon)):
            poly_start = polygon
            index_start = index2
            find_start = True
        if (not find_end & geom_end.within(polygon)):
            poly_end = polygon
            index_end = index2
            find_end = True
        if (find_start & find_end):
            break
    if poly_start != poly_end:
        df_mobike_per_discrit.loc[index_start, 'nb_mobike'] += 1
        df_mobike_per_discrit.loc[index_end, 'nb_mobike'] += 1
    else:
        df_mobike_per_discrit.loc[index_start, 'nb_mobike'] += 1
```
