```python
for index, name in df['Country (according to CN Gov):'].items():          
    cleanedName = name.split(' (')[0].strip()
    if cleanedName == "Thailiand":
        cleanedName = "Thailand"
    elif cleanedName == "The United Arab Emirates":
        cleanedName = "United Arab Emirates"
    elif cleanedName == "Brunei":
        cleanedName = "Brunei Darussalam"
    elif cleanedName == "Myanmar":
        cleanedName = "Burma"
    elif cleanedName == "East Timor":
        cleanedName = "Timor-Leste"
    elif cleanedName == "Kyrgyzstan":
        cleanedName = "Kyrgyz Republic"
    elif cleanedName == "United States of America":
        cleanedName = "United States"
    try:
        indicator = heritage_df[heritage_df['Country Name']==cleanedName]['2019 Score']
        indicator = heritage_df[heritage_df['Country Name']==cleanedName]['World Rank']
        df.at[index, 'HF Index of Economic Freedom (Scale: 1-100)'] = indicator
        df.at[index, 'Heritage Foundation Index of Economic Freedom (Global Ranking)'] = rank
    except ValueError:
        print('[critical]: no data for', cleanedName)
        continue
```

4. Moody's

We need ISO2 for moody's

```python
iso2 = {}
```

```python
for index, name in df['Country (according to CN Gov):'].items():    
    if name in iso2:
        continue        
    cleanedName = name.split(' (')[0].strip()
    if cleanedName == "Thailiand":
        cleanedName = "Thailand"
    if cleanedName == "The United Arab Emirates":
        cleanedName = "United Arab Emirates"
    url = 'http://www.restcountries.eu/rest/v2/name/'+cleanedName
    r = requests.get(url)
    if r.status_code == 200:
        data = r.json()
        country_code = data[0]['alpha2Code']
        iso2[name] = country_code
    else:
        print('[ERROR]['+ name +'] we cannot get the data, reason is ', r.status_code, r.text)
```

```python
df["isocode2"] = df["Country (according to CN Gov):"].apply(lambda x: iso2[x])
```

```python
ratings_df = pd.read_csv('data-ratings-2015.csv', delimiter=";")
ratings_df.head()
```

```python
ratings2015 = ratings_df[ratings_df['Year']==2015]
ratings2015.head()
```

```python
for index, isocode in df.isocode2.iteritems():
    try:
        moody = ratings2015[ratings2015['Country2Code']==isocode]['Moodys_NumericRating'].values[0]
        sp = ratings2015[ratings2015['Country2Code']==isocode]['SP_NumericRating'].values[0]
        fitch = ratings2015[ratings2015['Country2Code']==isocode]['Fitch_NumericRating'].values[0]
        df.at[index, "Moody's Socvereign Ratings List"] = moody
        df.at[index, 'S&P Sovereign Ratings List'] = sp
        df.at[index, 'Fitch Sovereign Ratings List'] = fitch
    except:
        print('[critical]: no data for', isocode)
        continue
```
