### 1. Remove the tags from HTML

```python
from bs4 import BeautifulSoup
import numpy as np
```

```python
corpus = []

def remove_html(text):
    return BeautifulSoup(text, 'lxml').get_text()
```

### 3. Remove the punctuations and digits

```python
import string #lib that contains every punctuations and digits
```

```python
#create a table for punctuations
translate_table_punc = dict((ord(char), u' ') for char in string.punctuation) #u is for unicode
#create a table for digits
translate_table_digits = dict((ord(char), None) for char in string.digits)

def remove_punctuation_and_digits(text):
    text = text.translate(translate_table_punc).lower() #remove punctuation
    text = text.translate(translate_table_digits) #remove digits
    return text 
```

```python
for i, v in gdf['description'].sample(n=200).dropna().iteritems():
    text = remove_html(v)
    text = remove_punctuation_and_digits(text)
    print(text)
```

### 4. Tokens

```python
from nltk.tokenize import casual
from nltk.corpus import stopwords
```

```python
def tokenize(text):
    return nltk.tokenize.casual_tokenize(text)
```

```python
for i, v in gdf['description'].sample(n=200).dropna().iteritems():
    text = remove_html(v)
    text = remove_punctuation_and_digits(text)
    words = tokenize(text)
    print(words)
```
